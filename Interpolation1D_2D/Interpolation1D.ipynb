```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as scp
from sklearn import metrics
```

```python
N = 5
def Sin(x):
    return np.sin(x)

def SinX1(x):
    return np.sin(1/x)

def signum(x):
    return np.sign(np.sin(8*x))
```

```python
x0 = np.linspace(-np.pi,np.pi,N)
y0 = [Sin(alfa) for alfa in x0]
y0_inverted = [SinX1(alfa) for alfa in x0]
y0_signum = [signum(alfa) for alfa in x0]
```

```python
_ = plt.plot(x0,y0, 'bo', label= " Sin(x)")
_ = plt.plot(x0,y0_inverted, 'ro', label="sin(1/x)")
_ = plt.plot(x0,y0_signum, 'co', label="sign(sin(8x))")
plt.legend(loc="lower right")
plt.show()
```

```python
inter_method = 'linear'
function_interpolated = scp.interp1d(x0,y0, kind= inter_method)
x_interp = np.linspace(x0.min(), x0.max(), 10*N)
y_interp = function_interpolated(x_interp)
y_true = Sin(x_interp)
print(f"MSE: {metrics.mean_squared_error(y_pred=y_interp, y_true=y_true):.8f}")
plt.plot(x0, y0, 'bo')
plt.plot(x_interp, y_interp, 'r-')
plt.grid(True)
```

```python
function_interpolated = scp.interp1d(x0,y0_inverted, kind= inter_method)
x_interp = np.linspace(x0.min(), x0.max(), 10*N)
y_interp = function_interpolated(x_interp)
y_true = Sin(x_interp)
print(f"MSE: {metrics.mean_squared_error(y_pred=y_interp, y_true=y_true):.8f}")
plt.plot(x0, y0_inverted, 'bo')
plt.plot(x_interp, y_interp, 'r-')
plt.grid(True)
```

```python
function_interpolated = scp.interp1d(x0,y0_signum, kind= inter_method)
x_interp = np.linspace(x0.min(), x0.max(), 10*N)
y_interp = function_interpolated(x_interp)
y_true = Sin(x_interp)
print(f"MSE: {metrics.mean_squared_error(y_pred=y_interp, y_true=y_true):.8f}")
plt.plot(x0, y0_signum, 'bo')
plt.plot(x_interp, y_interp, 'r-')
plt.grid(True)
```
